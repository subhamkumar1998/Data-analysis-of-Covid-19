# Graph to plot Most Active Cases 

```python
top10ActiveCases=df.groupby(by='Country/Region').max()[['Active','Date']].sort_values(by=['Active'],ascending=False).reset_index()
```

```python
fig=plt.figure(figsize=(16,10))
plt.title("Top 10 country with most Active Cases",size=22)
ax=sns.barplot(data=top10ActiveCases.iloc[:10],y='Active',x='Country/Region',linewidth=2,edgecolor='black')
plt.xlabel("Country")
plt.ylabel("Total Active Cases")
plt.show()
```

# To plot highest Number of Death 

```python
# top states with highest number of deaths
top10Deaths= df.groupby(by='Country/Region').max()[['Deaths','Date']].sort_values(by=['Deaths'],ascending=False).reset_index()


```

```python
fig=plt.figure(figsize=(16,10))
plt.title("Top 10 country with most Deaths",size=22)
ax=sns.barplot(data=top10Deaths.iloc[:10],y='Deaths',x='Country/Region',linewidth=2,edgecolor='black')
plt.xlabel("Country/Region")
plt.ylabel("Total Deaths")
plt.show()
```
